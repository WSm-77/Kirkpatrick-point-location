# This file contains showcase of Kirkpatrick's point location algorithm 

```python
import json
import os
from  point_location.visual_kirkpatrick.visual_kirkpatrick import VisualKirkpatrick
```

```python
test_no = 1
test_dir = "test_input"
test_name = f"test{test_no}.json"

with open(os.path.join(test_dir, test_name), 'r') as file:
    data = json.load(file)

vertices = data["vertices"]
segments = data["segments"]

kirkpatrick = VisualKirkpatrick(vertices, segments)

kirkpatrick.visual_preprocessing()
```

```python
points_to_locate = data["points_A"]
points_iter = iter(points_to_locate)
```

```python
kirkpatrick.visual_locate_point((2.0, 0.5))
```

```python
kirkpatrick.visual_locate_point(next(points_iter))
```
