# This file contains showcase of Kirkpatrick's point location algorithm 

```python
import json
import os
from  point_location.visual_kirkpatrick.visual_kirkpatrick import VisualKirkpatrick
from point_location.visual_kirkpatrick.showcase_utils import pick_point, create_planar_subdivision
import matplotlib.pyplot as plt
```

```python
test_no = 4
test_dir = "test_input"
test_name = f"test{test_no}.json"

with open(os.path.join(test_dir, test_name), 'r') as file:
    data = json.load(file)

vertices = data["vertices"]
segments = data["segments"]
```

```python
%matplotlib inline
kirkpatrick = VisualKirkpatrick(vertices, segments)
kirkpatrick.visual_preprocessing()
```

```python
points_to_locate = data["points_A"]
points_iter = iter(points_to_locate)
```

```python
kirkpatrick.visual_locate_point(next(points_iter))
```

```python
%matplotlib ipympl
points, edges = create_planar_subdivision()
```

```python
%matplotlib inline
kirkpatrick = VisualKirkpatrick(points, edges)
kirkpatrick.visual_preprocessing()
```

```python
%matplotlib ipympl
point = pick_point(kirkpatrick)
```

```python
%matplotlib inline
plt.close("all")
kirkpatrick.visual_locate_point(point[0])
```
